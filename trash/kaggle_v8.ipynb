# Import

```python
import numpy as np
import pandas as pd
import csv
import datetime
import math
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.model_selection import RandomizedSearchCV
import lightgbm as lgbm
```

# My Library

```python
#製造answer.csv的第1列
def answer_header():
    header = []
    for i in range(28):
        header.append('time_slot_' + str(i) )
    return header
```

# Read File

```python
reference = pd.read_csv('reference.csv')
answer_v2 = pd.read_csv('answer_v2.csv')
the_labels = pd.read_csv('public/The_Labels.csv')
```

# Main

```python
X = reference[answer_header()]
params = {'max_depth': 5,
         }
y = []
model = []
for i in range(0,28):
    y.append(the_labels[answer_header()]['time_slot_' + str(i)])
    train_set = lgbm.Dataset(X, y[i], silent=True)
    model.append(lgbm.train(params, train_set=train_set, num_boost_round=30))
    print(i)
```

```python
with open('answer_v8.csv', 'w') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['user_id'] + answer_header())
    for i in range(57159, 94251):
        data = answer_v2[answer_v2['user_id'] == i][answer_header()]
        results = [i]
        for j in range(28):
            result = model[j].predict(data, num_iteration=model[j].best_iteration)[0]
            results.append(result)
        writer.writerow(results)
        print(i)
```
